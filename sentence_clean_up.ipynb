### Clean up the negative-replaced-sentences.
trimm the sentence by only taking the words between the first and the last occurance of 'chem'

```python
import sys
sys.path.append("./packages") # append to system path
import data_preparation as dp

import csv

trimmed_sentence = []
with open('./data/identified_chemical_negative.csv','rb') as myfile:
    thisReader = csv.reader(myfile)
    for eachLine in thisReader:
        this_token = dp._tokenize_sentence(eachLine)
        this_trimmed_token = dp._trimming(this_token, buffer = 8)
        trimmed_sentence.append(this_trimmed_token)

# write to file
with open('./data/trimmed_indentified_chemical_negative.csv','wb') as myfile:
    thisWriter = csv.writer(myfile)
    for eachSen in trimmed_sentence:
        thisWriter.writerow(eachSen)
```
